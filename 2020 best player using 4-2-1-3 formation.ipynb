# Best Goal Keeper Analysis

```python
#this is the rate of how good a player must be on each ability. If it is d * attacking_speed for winger, it means the winger must have a good attacking speed. If it is a*defending for winger, it means the winger does not need to be good at defending
a = 0.5
b = 1
c = 2
d = 3
```

```python
#GoalKeeping Characteristics
df['Shot_Stopper'] = (b*df.movement_reactions + b*df.mentality_composure + a*df.gk_speed + a*df.power_strength + c*df.power_jumping + b*df.goalkeeping_positioning + c*df.goalkeeping_diving + d*df.goalkeeping_reflexes + b*df.goalkeeping_handling)/(2*a + 4*b + 2*c + 1*d)
df['Sweeper'] = (b*df.movement_reactions + b*df.mentality_composure + a*df.gk_speed + b*df.attacking_short_passing + a*df.skill_long_passing + b*df.power_jumping + b*df.goalkeeping_positioning + b*df.goalkeeping_diving + d*df.goalkeeping_reflexes + b*df.goalkeeping_handling + d*df.goalkeeping_kicking + c*df.mentality_vision)/(2*a + 7*b + 1*c + 2*d)
```

```python
plt.figure(figsize=(15,6))

#Generate sequential data and plot
sd1 = df.sort_values("Shot_Stopper", ascending=False)[:7]
x1 = np.array(list(sd1['long_name']))
y1 = np.array(list(sd1['Shot_Stopper']))

sns.barplot(x1,y1, palette='colorblind')
plt.ylabel("Shot Stopper Score")
plt.xlabel("Player Name")
# plt.savefig('shot stopper.png') # you can uncomment this if you want to save the result.
```

```python
plt.figure(figsize=(15,6))

#Generating sequential data and plot
sd1 = df.sort_values("Sweeper", ascending=False)[:7]
x1 = np.array(list(sd1['long_name']))
y1 = np.array(list(sd1['Sweeper']))

sns.barplot(x1,y1, palette='colorblind')
plt.ylabel("Sweeper Score")
plt.xlabel("Players Name")
print(x1[0],"is the best GOAL KEPPER in 2020")
```

# THE FIRST FOUR BEST DEFENDERS FOR 2020 

```python
#Choosing Defenders based on a characteristics that must be possessed by a defender
df['centre_backs'] = (d*df.movement_reactions +c*df.mentality_interceptions + d*df.defending_sliding_tackle + d*df.defending_standing_tackle + b*df.mentality_vision + b*df.mentality_composure + b*df.attacking_crossing  + a*df.attacking_short_passing + b*df.skill_long_passing + c*df.movement_acceleration + b*df.movement_sprint_speed + d*df.power_stamina + d*df.power_jumping + d*df.attacking_heading_accuracy + b*df.power_long_shots + d*df.defending_marking + c*df.mentality_aggression)/(6*b + 3*c + 7*d)
df['Wing_Backs'] = (b*df.skill_ball_control + a*df.skill_dribbling + a*df.defending_marking + d*df.defending_sliding_tackle + d*df.defending_standing_tackle + a*df.mentality_positioning + c*df.mentality_vision + c*df.attacking_crossing  + b*df.attacking_short_passing + c*df.skill_long_passing + d*df.movement_acceleration + d*df.movement_sprint_speed + c*df.power_stamina + a*df.attacking_finishing)/(4*a + 2*b + 4*c + 4*d)
```
